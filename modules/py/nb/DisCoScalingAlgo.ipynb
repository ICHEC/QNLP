```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import scipy as cp

from mpl_toolkits.mplot3d import Axes3D
```

```python
N = 8 #Number of basis vectors
M = 8 #Number of comparison vectors
```

```python
def q_scale(N, M):
    return np.sqrt(N*M)*np.log(M)
```

```python
def cl_scale(N,M):
    return N*M
```

```python
x = np.array(range(1,32))
y = np.array(range(1,32))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.plot(xs=x, ys=y, zs=q_scale(x,y), label="Quantum")
plt.plot(xs=x, ys=y, zs=cl_scale(x,y), label="Classical")
plt.legend()
ax.set_xlabel("Num. data vectors")
ax.set_ylabel("Num. comparison vectors")
ax.set_zlabel("Num. queries")
```

```python
fig = plt.figure()
plt.plot(x, q_scale(x,2), label="Quantum (2 vec. comp.)")
plt.plot(x, q_scale(x,8), label="Quantum (8 vec. comp.)")
plt.plot(x, cl_scale(x,2), label="Classical (2 vec. comp.)")
plt.plot(x, cl_scale(x,8), label="Classical (8 vec. comp.)")
plt.legend()
plt.xlabel("Num. data vectors")
#ax.set_ylabel("Num. comparison vectors")
plt.ylabel("Num. queries")
```
